```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

df = pd.read_csv('/Users/Sishe/Desktop/ProjectGlassdoor/types.csv', names=['Company', 'Size', 'Industry', 'SectorType', 'Skills', 'Salary', 'Location', 'Rating', 'Title'])
df = df.drop(df.index[[0,]])#drop first row 
df['City'], df['State'] = df['Location'].str.split(',', 1).str # Split into state and city
df['Python'] = df['Skills'].str.contains('Python')
df['Excel'] = df['Skills'].str.contains('Excel'or'excel')
df['SQL'] = df['Skills'].str.contains('Sequel'or'SQL'|'sql')
df['R Shiny'] = df['Skills'].str.contains('R'or 'Shiny')
df['Tableau'] = df['Skills'].str.contains('Tableau'or'tableau')
df['MySQL'] = df['Skills'].str.contains('MySQL' or 'Sequel')
df['Hive'] = df['Skills'].str.contains('Hive')
df['SAS'] = df['Skills'].str.contains('Sas')
df['Hive'] = df['Skills'].str.contains('Hive')
df['Hadoop'] = df['Skills'].str.contains('Hadoop' or 'HADOOP')
df['Company'] = df['Company'].str.strip()
df = df.drop(['Salary','Location'], axis=1)
df['Sector'] = np.where(df['SectorType'].str.contains('Public'),
        'Public', np.where(df['SectorType'].str.contains('Private'),'Private',
        np.where(df['SectorType'].str.contains('NonProfit'),'NonProfit',
                 np.where(df['SectorType'].str.contains('College'),'CollegeUniversity',
                          np.where(df['SectorType'].str.contains('Government'),'Government',
             'SubsidiaryorBusiness Segment')))))

```

```python

df1 = pd.read_csv('/Users/Sishe/Desktop/ProjectGlassdoor/infoglassdoor.csv', names=['salary', 'Company', 'Industry', 'SectorType'],encoding='latin-1')
df1 = df1.drop(df1.index[[0,]])
df1['Company'], df1['state'] = df1['Company'].str.split('äóñ', 1).str
df1 = df1.drop(['Industry','SectorType', 'state'], axis=1)
df1['Company'] = df1['Company'].str.strip()
```

```python
dataAnalyst = pd.merge(df1, df, on='Company', how = 'left')
dataAnalyst['Low-Salary'], dataAnalyst['High-Salary'] = dataAnalyst['salary'].str.split('-', 1).str
```

```python

#dataAnalyst = dataAnalyst.drop('salary', axis=1)

dataAnalyst['High-Salary'] = dataAnalyst['High-Salary'].replace({'\$':''}, regex = True)
dataAnalyst['Low-Salary'] = dataAnalyst['Low-Salary'].replace({'\$':''}, regex = True)
dataAnalyst['High-Salary'] = dataAnalyst['High-Salary'].str.replace('Est. Salary' and 'k', '')
dataAnalyst['Low-Salary'] = dataAnalyst['Low-Salary'].str.replace('Est. Salary', '')





```
